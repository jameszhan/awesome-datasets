```python
import pandas as pd
import faker
```

```python
f = faker.Faker('zh-CN')
```

```python
num = 20
df = pd.DataFrame({
    '客户名称': [f.name() for i in range(num)],
    '年龄': [f.random_int(25, 40) for i in range(num)],
    '最后去电时间': [f.date_between(start_date='-1y', end_date='today').strftime('%Y年%m月%d日') for i in range(num)],
    '意向':  [f.random_element(('有','无')) for i in range(num)],
    '地址': [f.street_address() for i in range(num)]
})
df
```
