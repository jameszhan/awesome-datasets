```python
import numpy as np

def ewma(series, alpha):
    y = np.zeros_like(series)
    y[0] = series[0]
    for i in range(1, len(series)):
        y[i] = alpha * series[i] + (1 - alpha) * y[i-1]
    return y
```

```python
alpha = 0.07995558537067671
series = np.ones(10) * 10

ewma(series, alpha)
```

```python
series = np.arange(20) * 10

ewma(series, alpha)
```

```python
series = np.ones(300, dtype=np.float64) * 10
series[0:5] = 0
series[90:180] = 0

ewma(series, alpha)
```
